# Wprowadzenie

### Złoto jest metalem szlachetnym o niezwykłych właściwościach fizycznych oraz chemicznych. Jest jedynym metalem monetarnym. Stanowi składnik rezerw walutowych w bankach oraz jest prywatnym aktywem inwestycyjnym. W przeciwieństwie do innych surowców bieżące wydobycie złota znaczący wpływ na jego cenę. Złoto jest jednym z najpopularniejszych towarów na rynku terminowym.

# Dane statystyczne

### Dane, które zostały wykorzystane do analizy, pochodzą z bazy danych strony „Datahub”.Strona posiada tysiące zestawów danych, z których korzystają miedzy innymi „Amnesty international” czy University of Oxford. Oryginalne dane ze strony Datahub, zawierają roczne ceny złota od 1950 do 2020 roku oraz miesięcznie ceny złota, również od 1950 do 2020 roku.

# Analiza - Wstęp

### Do analizy wykorzystałem miesięczne ceny złota w latach 2011-2020. Ceny złota wyrażone są w dolarach amerykańskich. Otrzymane wyniki obliczeń zaokrągliłem. Średnia arytmetyczna, kwartyle oraz dominanta zastały zaokrąglone do liczb całkowitych. Reszta obliczeń została zaokrąglona do drugiego miejsca po przecinku.
